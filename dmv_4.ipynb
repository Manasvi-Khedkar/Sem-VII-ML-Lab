## Data Visualization using matplotlib

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('AQI - AQI.csv')

# Display the first few rows to explore the dataset
df.head()
```

```python
# 2. Explore the dataset
df.info()
```

```python
df.describe()
```

```python
# 3. Identify relevant variables
# Check for relevant columns
df.columns
```

```python
import matplotlib.pyplot as plt
df['date'] = pd.to_datetime(df['date'], format='mixed', dayfirst=False, errors='coerce')
df.head()
```

```python
#4. Create line plots or time series plots to visualize the overall AQI trend over time
# Line plot for AQI trend over time

plt.plot(df['date'], df['rspm'], color='blue', label='AQI (RSPM)')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title('AQI Trend Over Time')
plt.legend()
plt.show()
```

```python
# 5)Plot individual pollutant levels (e.g., PM2.5, PM10, CO) on separate line plots to 
#visualize their trends over time
# PM2.5 trend over time
plt.plot(df['date'], df['pm2_5'], color='orange', label='PM2.5')

plt.xlabel('Date')
plt.ylabel('PM2.5 Level')
plt.title('PM2.5 Trend Over Time')
plt.legend()
plt.show()

# Similar code can be used for SO2 and NO2 levels
```

```python
plt.xlabel('Date')
plt.ylabel('so2 Level')
plt.title('so2 Trend Over Time')
plt.show()
plt.plot(df['date'], df['so2'], label='so2', color='green')
```

```python
plt.xlabel('Date')
plt.ylabel('no2 Level')
plt.title('no2 Trend Over Time')
plt.legend()
plt.show()
plt.plot(df['date'], df['no2'], label='no2', color='red')
```

```python
# 8. Use scatter plots or bubble charts to explore the relationship between AQI values and 
#pollutant levels
# Scatter plot for AQI and PM2.5 levels
plt.figure(figsize=(10, 6))
plt.scatter(df['pm2_5'], df['rspm'], color='green', alpha=0.5)
plt.xlabel('PM2.5')
plt.ylabel('AQI (RSPM)')
plt.title('Relationship between PM2.5 and AQI')
plt.show()
```
